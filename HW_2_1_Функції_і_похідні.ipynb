1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
x = sp.symbols('x')

f = 10 * x - 5
derivative_f = sp.diff(f, x)

print(f'Derivative of function 10x - 5:  {derivative_f}')
```

```python
def f(x):
    return(10 * x - 5)

x = np.linspace(0, 10, 100)
plt.plot(x, f(x), color='g', label='function')
plt.hlines(10, 0, 10, color='r', label='derivative')
plt.legend()
plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')

f_sin = sp.sin(x)
f_ln = sp.ln(x)
def derivative_f(f):
    return sp.diff(f, x)

print(f'Derivative of function sin(x):  {derivative_f(f_sin)}')
print(f'Derivative of function ln(x):  {derivative_f(f_ln)}')
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 10001)

def ln(x, diff=False):
    with np.errstate(divide='ignore', invalid='ignore'):
        if diff:
            return 1/x
        else:
            return np.log(x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))
ax1.plot(x, np.sin(x), color='g', label='function sin(x)')
ax1.plot(x, np.cos(x), '--', color='r', label='derivative cos(x)')
ax1.legend(loc='upper right')

ax2.plot(x, ln(x), color='g', label='function ln(x)')
ax2.plot(x, ln(x, diff=True), '--', color='r', label='derivative 1/x')
ax2.set_ylim(-5, 5)
ax2.legend(loc='upper right')

plt.show()
```
